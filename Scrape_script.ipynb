```python
import requests

def get_all_repo_urls(org_name):
    base_url = f"https://api.github.com/orgs/{org_name}/repos"
    headers = {
        "Accept": "application/vnd.github.v3+json",
        "Authorization": "Bearer <github pat>"
    }
    repo_urls = []
    try:
        page = 1
        while True:
            response = requests.get(base_url, headers=headers, params={"page": page})
            response.raise_for_status()

            repos = response.json()
            if not repos:
                break

            for repo in repos:
                repo_urls.append(repo["html_url"])

            page += 1
    except requests.exceptions.RequestException as e:
        print(f"Error occurred: {e}")
        return None

    return repo_urls

```

```python
import os
import requests
import shutil
import json

count=1
def download_markdown_files(repo_url, output_dir,headers):
    global count

    response = requests.get(repo_url,headers=headers)
    if response.status_code != 200:
        print(f"Failed to fetch repository contents. Status code: {response.status_code}")
        return

    contents_data = response.json()

    # Loop through all files and download markdown files
    for content in contents_data:
        if content['type'] == 'file' and content['name'].endswith('.md'):
            file_url = content['download_url']
            files=os.listdir(output_dir)
            if content['name'] in files:
              new_name=content['name'].split(".")[0]+str(count)+".md"
              count+=1
              file_name = os.path.join(output_dir, new_name)
            else:
              file_name = os.path.join(output_dir, content['name'])
            download_file(file_url, file_name)

        elif content['type'] == 'dir':
            new_url=repo_url+"/"+content['name']
            download_markdown_files(repo_url=new_url, output_dir=output_dir,headers=headers)

def download_file(url, file_path):
    #Download the file and save it to the specified file path
    response = requests.get(url)
    if response.status_code == 200:
        with open(file_path, 'wb') as file:
            file.write(response.content)
        print(f"Downloaded: {file_path}")
    else:
        print(f"Failed to download: {file_path}. Status code: {response.status_code}")

if __name__ == "__main__":
    orgs=["redhat-performance", "cloud-bulldozer", "arcalot", "perftool-incubator", "redhat-chaos", "distributed-system-analysis", "Appservices-perfscale"]
    all_repo_urls=[]
    for i in orgs:
      org_urls=get_all_repo_urls(i)
      all_repo_urls=all_repo_urls+org_urls
    all_repo_urls=list(map(lambda x: x.replace("github.com", "api.github.com/repos")+"/contents",all_repo_urls))
    # repository_url = "https://github.com/cloud-bulldozer/kube-burner"
    # api_url = repository_url.replace("github.com", "api.github.com/repos")
    # print(api_url)
    print(all_repo_urls,sep="\n")
    output_directory = "output"
    headers = {
        "Accept": "application/vnd.github.v3+json",
        "Authorization": "Bearer <github_pat>"
    }
    #print(all_repo_urls)
    os.makedirs(output_directory, exist_ok=True)
    for i in all_repo_urls:
      print(i)
      download_markdown_files(i, output_directory,headers)
```
